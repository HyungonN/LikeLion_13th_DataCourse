```python
dat = np.random.rand(100,4) #랜덤 data_set 발생
df = pd.DataFrame(dat, columns = "A B C D".split())
df.head()
```

```python
df['A'] #DataFrame은 Series로 이루어져 있다.

df[['A','C']].iplot(kind='line', xTitle = "데이터idx", yTitle = "값",title = "랜덤데이터프레임")
```

```python
df.loc[:, ['A','C']].iplot(kind='line', xTitle = "데이터idx", yTitle = "값",title = "랜덤데이터프레임")
```

```python
df.iplot()
```

```python
df.iplot(kind="scatter",x="A", y = "B", mode = "markers", size = 20)
```
