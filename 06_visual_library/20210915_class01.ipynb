```python
plt.figure(figsize=(8,6))
sns.barplot(x="day", y="total_bill", data=tips)
plt.show() 
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show() 
```

```python
sns.boxplot(x="day", y = "tip", data = tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='day', y='size', data=tips)
plt.show()
```

```python
sns.boxplot(x='sex', y='tip', data=tips)
plt.show()
```

```python
sns.boxplot(x='day', y='tip', hue='sex', data=tips)
plt.show()
```

```python
sns.boxplot(x="day", y="tip", hue="smoker", data = tips)
```

```python
sns.boxplot(x="day", y="tip", hue="time", data = tips)
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, size=7)
plt.show
```

```python
fg = sns.load_dataset("flights")
fg
```

```python
sns.scatterplot(x="year", y="passengers", data=fg)
```

```python
sns.scatterplot(x="year", y="passengers", hue="month", data=fg)
```

```python
fg.columns
```

```python
fgpivot = fg.pivot("month", "year", "passengers")
fgpivot
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(fgpivot, annot=True,fmt="d")
plt.show()
```
