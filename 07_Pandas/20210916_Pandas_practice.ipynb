```python
data = sns.load_dataset("titanic")
data
```

```python
sns.histplot(data.fare)
```

```python
data.fare.describe()
```

```python
### min~25% :1번그룹
#    25%~50% : 2번그룹
#        ...
#
```

```python
data_fare01 = data.loc[ (data.fare >=0) & (data.fare < 7.9104) , :   ]
data_fare02 = data.loc[ (data.fare >=7.9104) & (data.fare < 14.4542) , :   ]
data_fare03 = data.loc[ (data.fare >=14.4542) & (data.fare < 31) , :   ]
data_fare04 = data.loc[ (data.fare >=31), :   ]
```

```python
sns.histplot(data_fare01["fare"])
```

```python
data.loc[ (data.fare >=0) & (data.fare < 7.9104) , "group_fare"    ] =1
data.loc[ (data.fare >=7.9104) & (data.fare < 14.4542) , "group_fare"] =2
data.loc[ (data.fare >=14.4542) & (data.fare < 31) , "group_fare"   ] =3
data.loc[ (data.fare >=31), "group_fare"   ]=4
data
```

```python
data.loc[ data.age < 10 , "tens"] = 0
data.loc[ (data.age >= 10) & (data.age<20) , "tens" ] = 1 
data.loc[ (data.age >= 20) & (data.age<30) , "tens" ] = 2 
data.loc[ (data.age >= 30) & (data.age<40) , "tens" ] = 3 
data.loc[ (data.age >= 40) & (data.age<50) , "tens" ] = 4 
data.loc[ (data.age >= 50) & (data.age<60) , "tens" ] = 5 
data.loc[ (data.age >= 60) & (data.age<70) , "tens" ] = 6 
data.loc[ (data.age >= 70) & (data.age<80) , "tens" ] = 7 
data.loc[ (data.age >= 80) , "tens" ] = 8 
```
