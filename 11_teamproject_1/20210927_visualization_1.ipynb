```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
shdata = pd.read_excel("Shinhancard.xlsx")
shdata.head()
```

```python
#"M013","M014","M015","M016","M001","M002","M003","M004","M005","M024","M025","M026","M028","M029"

M001 = shdata.loc[ shdata["업종"]== "M001_한식", :]
M002 = shdata.loc[ shdata["업종"]== "M002_일식/중식/양식", :]
M003 = shdata.loc[ shdata["업종"]== "M003_제과/커피/패스트푸드", :]
M004 = shdata.loc[ shdata["업종"]== "M004_기타요식", :]
M005 = shdata.loc[ shdata["업종"]== "M005_유흥", :]
d1 = pd.merge(M001,M002, how = "outer")
d2 = pd.merge(M003,M004, how = "outer")
d3 = pd.merge(d1,d2, how = "outer")
food_sh = pd.merge(d3,M005, how = "outer")
food_sh.info()
food_sh.head()
```

```python
food_sh.to_csv("shinhan_Food.csv",index=False)
```
