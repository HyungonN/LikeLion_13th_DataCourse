```python
def page_read() :
    page = driver.page_source
    soup = BeautifulSoup(page, "html.parser")
    
    name = soup.find_all("b")
    address_set = []
    base_xpath = '//*[@id="container"]/div/div[3]/ul/li['
    end_xpath = ']/a/div/ul/li[1]'
    for i in range(1,13,1) :
        one_xpath = base_xpath + str(i) +end_xpath
        data = driver.find_element_by_xpath(one_xpath)
        address_set.append(data.text)
    
    page_dic = {}
    for i in range(12) :
        page_dic[name[i].text] = address_set[i]
    
    return page_dic
```

```python
import time
welfare_loc = {}
for page_num in range(4,14,1) :
    welfare_loc.update(page_read())
    page_button = driver.find_element_by_xpath(
        '//*[@id="container"]/div/div[4]/a['+str(page_num)+']')
    page_button.click()
    time.sleep(3)
```
