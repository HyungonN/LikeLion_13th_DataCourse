```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
```

```python
driver = webdriver.Chrome("./chromedriver")

url = "https://coronaboard.kr"
driver.get(url)

time.sleep(3)
#driver.implicity_wait(3) : 데이터 모두 로드될때까지 기다림
```

```python
#더보기 클릭
from selenium.webdriver import ActionChains

some_tag = driver.find_element_by_id("show-more")
ActionChains(driver).move_to_element(some_tag).perform()
some_tag.click()
some_tag = driver.find_element_by_id('show-more')
ActionChains(driver).move_to_element(some_tag).perform()
some_tag.click()
```

```python
#표에서 보여줄것 선택
some_tag = driver.find_element_by_xpath('//*[@id="global-slide"]/div/div[2]/div/div/button/div/div/div')
ActionChains(driver).move_to_element(some_tag).perform()
some_tag.click()
```

```python
sel_menu = driver.find_element_by_xpath('//*[@id="bs-select-1-1"]')
sel_menu.click()
sel_menu = driver.find_element_by_xpath('//*[@id="bs-select-1-5"]')
sel_menu.click()
```

```python
sel_country = driver.find_elements_by_xpath('//*[@id="country-table"]/div/div/table/tbody/tr/td[2]')
len(sel_country)
for one in sel_country :
    print(one.text)
```

```python
# 어제 : //*[@id="global-slide"]/div/div[2]/ul/li[2]/a

flag_today = False # False : 어제, Today : 오늘
if flag_today==False:
    sel_yesterday = driver.find_element_by_xpath('//*[@id="global-slide"]/div/div[2]/ul/li[2]/a')
    sel_yesterday.click()
    
all_data = []
for i in range(2, 11):
    tmp = '//*[@id="country-table"]/div/div/table/tbody/tr/td[{}]'.format(i)   #xpath따라서 data수집
    # print(tmp)
    sel_ele = driver.find_elements_by_xpath(tmp)
    
    
    column_data = []
    for one in sel_ele:
        # print(one.text)
        column_data.append(one.text)
    
    print( len(sel_ele) )
    all_data.append(column_data)
    print(column_data)
    print()
    
all_data
```

```python
# 데이터프래임 만들기
import pandas as pd

dict_dat = { "국가":all_data[0], 
             "확진자":all_data[1],
             "위중증":all_data[2],
             "사망자":all_data[3],
             "완치":all_data[4],
             "치명(%)":all_data[5],
             "완치(%)":all_data[6],
             "발생률":all_data[7],
             "인구수":all_data[8]
           }

dat = pd.DataFrame(dict_dat)
dat
```

```python
dat["완치자_합계"] = dat["완치"].str.split("\n").str[0]
dat["완치자_1일"] = dat["완치"].str.split("\n").str[1]
dat.head()
```

```python
dat = dat.drop( ["확진자","사망자","완치"], axis = 1 )
```

```python
dat['위중증'] = dat['위중증'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat['발생률'] = dat['발생률'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['인구수'] = dat['인구수'].str.replace(pat=r'[,]', repl=r'', regex=True)
```

```python
dat['확진자_합계'] = dat['확진자_합계'].str.replace(pat=r'[,]', repl=r'', regex=True)
dat['확진자_1일'] = dat['확진자_1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat.head()
```

```python
dat['사망자_합계'] = dat['사망자_합계'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat['사망자_1일'] = dat['사망자_1일'].str.replace(pat=r'[,()]', repl=r'', regex=True)
dat.head()
```

```python
dat["완치자_합계"] = dat["완치자_합계"].str.replace(pat=r'[,]', repl=r'', regex=True)
dat["완치자_1일"] = dat["완치자_1일"].str.replace(pat=r'[ ,()+ ]', repl=r'', regex=True)

dat.head()
```

```python
dat["위중증_합계"] = dat["위중증"].str.split("\n").str[0]
dat["위중증_1일"] = dat["위중증"].str.split("\n").str[1]
dat = dat.drop(["위중증"], axis=1)
dat.head()
```
