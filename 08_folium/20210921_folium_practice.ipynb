```python
import pandas as pd
import folium
```

```python
dat = pd.read_csv("seoul_public_toilet.csv", encoding="cp949")
dat
```

대명칭 위도 경도(WSG84X좌표	WSG84Y좌표) 가져오기

```python
dat.columns
```

```python
dat_tmp = dat.loc[0:100, ["대명칭","WSG84X좌표","WSG84Y좌표"]]
dat_tmp
```

```python
dat_tmp.columns = ["name", "경도", "위도"]
dat_tmp
```

```python
dat_tmp.shape
```

```python
des = dat_tmp["name"]
dat_p1 = list(dat_tmp["위도"])
dat_p2 = dat_tmp["경도"]
```

```python
dat_xy = list(zip(dat_p1,dat_p2))
```

```python
dat_xy
```

```python
import time 
time_start = time.time()

m = folium.Map(location = [37.55578700870461, 126.93996768756287], zoom_start = 13)
for name, loc in zip(des,dat_xy) :
  folium.Marker(location = loc, 
                tooltip= name, 
                icon=folium.Icon(color="blue", icon = "info-sign")).add_to(m)
m.save("seoul_toliet_100.html")

time_end = time.time()

print(time_end - time_start)
```

```python
import pandas as pd
import folium

data = pd.read_csv("제주도장소데이터.csv", encoding="cp949")
data.columns
```

```python
is_land = (data["구분"] == "지명관련")  #조건부 Pandas 데이터 뽑기
jejuland = data[is_land]
jejuland
```

```python
#데이터 100개만 추출 & columns 이름 변경
jejuland_tmp = jejuland.loc[0:100, ['장소명','위치좌표 X축값 ', '위치좌표 Y축값 ']]
jejuland_tmp.columns = ["name", "경도", "위도"]
```

```python
#
des = jejuland_tmp["name"]
dat_p1 = list(jejuland_tmp["위도"])
dat_p2 = jejuland_tmp["경도"]
dat_xy = list(zip(dat_p1,dat_p2))


import time #계산시간 표시
time_start = time.time()

jejumap = folium.Map(location = [33.435, 126.5], zoom_start = 10)
for name, loc in zip(des,dat_xy) :
  folium.Marker(location = loc, 
                tooltip= name, 
                icon=folium.Icon(color="blue", icon = "info-sign")).add_to(jejumap)
jejumap.save("제주지형_100.html")

time_end = time.time()

print(time_end - time_start)
```
